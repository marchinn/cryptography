```python
from collections import Counter
import matplotlib.pyplot as plt
import re

text = """GB OR, BE AEG GB OR: GUNG VF GUR DHRFGVBA:
JURGURE VVF ABOYRE VA GUR ZVAQ GB FHSSRE
GUR FYVATF NAQ NEEBJF BS BHGENTRBHF SBEGHAR,
BE GB GNXR NEZF NTNVAFG N FRN BS GEBHOYRF,
NAQ OL BCCBFVAT RAQ GURZ? GB OVR: GB FYRRC;
AB ZBER; NAQ OL N FYRRC GB FNL JR RAQ
GUR URNEG-NPUR NAQ GUR GUBHFNAQ ANGHENY FUBPXF
GUNG SYRFU VF URVE GB, WF N PBAFHZZNGVBA
QRIBHGYL GB OR JVFU'Q, GB QVR, GB FYRRC;
GB FYRRC: CREPUNAPR GB QERNZ: NL, GURERT GUR EHO;
SBE VA GUNG FYRRC BS QRNGU JUNG QERNZF ZNL PRZR
JURA JR UNIR FUHSSYRQ BSS GUVF ZBEGNY PBVY,
ZHFG TVIR HF CNHFR: GURERT GUR ERFCRPG
GUNG ZNXRF PNYNZVGL BS FB YBAT YVSR;"""
text = text.lower()
text_copy = re.sub(r"[^\w\s]", "", text)
words = text_copy.split()
text_copy = text_copy.replace(" ", "")
letter_freq = Counter(text_copy)

word_freq = Counter(words)

plt.bar(letter_freq.keys(), letter_freq.values())
plt.xlabel("Буквы")
plt.ylabel("Частота")
plt.title("Частотный анализ букв")
plt.show()

plt.figure(figsize=(12, 5))
plt.bar(word_freq.keys(), word_freq.values())
plt.xlabel("Слова")
plt.xticks(rotation=90)
plt.ylabel("Частота")
plt.title("Частотный анализ слов")
plt.show()

# проверка k=13
key = "abcdefghijklmnopqrstuvwxyz"
test = text.lower()
result = ""
for letter in test:
    index = key.find(letter)
    if index == -1:
      result += letter
    else:
      new_index = (index + 13) % len(key)
      result += key[new_index]
result = result.upper()
print (result)
```
