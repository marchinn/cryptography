```python
import math
import random

class ElGamal:
  def __init__(self, p, q):
    self.p = p #выбрали число
    self.q = q

  def do(self, m):
    c = random.randint(2,self.p-2) # выбрали своё секретное число с..
    d = self.mod_pow(self.q, c, self.p) # .. и вычислили открытое число d

    # отправитель ABC ---m---> BCD
    if m >= self.p: # проверяемм
      print ("it's not coorect! m>=p!")
      pass

    k = random.randint(1,self.p-2) # если всё ок, ABC формирует k..
    R = self.mod_pow(self.q, k, self.p) # ..и вычисляет числа R и E
    E = m * self.mod_pow(d, k, self.p)

    # снова передачка
    # ABC ---R,E---> BCD

    m1 = (E * self.mod_pow(R, self.p - 1 - c, self.p)) % self.p # BCD рассекречивает сообщение

    print("c ", c)
    print("d ", d)
    print("k ", k)
    print("R ", R)
    print("E ", E)

    self.check(m, m1)

  def check(self, m, m1):
    if m == m1:
      print ("match! - ", m, ", ", m1)
    else:
      print("messages don't match: ", m, ", ", m1)

  def mod_pow(self, base, exponent, modulus):  # алгоритм быстрового возведения в степень
    result = 1 # начинаем с самой масенькой степени, рассмотрим один цикл в mod_pow(g=7, a=9, p=41) или g^a % p
    base %= modulus # понижаем степень, чтобы она не росла слишком быстро т.е на 1ом шаге 7(g) % 9(a) = 7
    while exponent > 0:
        if exponent % 2 == 1:  # если степень нечётная то
           result = (result * base) % modulus # т.е 1*7 % 9 = 7
        exponent //= 2 #  если чётное то понижаем вдвое, используем целочисленное деление
        base = (base * base) % modulus # присваиваем основанию ((основание возведённое в степень самого себя) % 9) и тд
    return result # на выходе, когда степень стала <= 0, возвращаем полученное число

p = 139
q = 70
message = 53

E = ElGamal(p, q)
E.do(message)
```
